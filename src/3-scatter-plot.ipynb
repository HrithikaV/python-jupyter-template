### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.figure(figsize=(6, 4))  # Set figure size
plt.scatter(x, y, color='blue', marker='o', s=100, edgecolors='black')  # Scatter plot

# Annotate each point with its (x, y) value
for i in range(len(x)):
    plt.text(x[i], y[i], f'({x[i]}, {y[i]})', fontsize=10, ha='right', va='bottom')

# Customize the plot
plt.title("Scatter Plot with Annotations", fontsize=14)  # Add title
plt.xlabel("X Values", fontsize=12)  # Label for x-axis
plt.ylabel("Y Values", fontsize=12)  # Label for y-axis
plt.grid(True, linestyle='--', alpha=0.6)  # Add grid

# Show the plot
plt.show()
```
