### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)  # 100 points for a smooth curve
y = x**2  # Square of x

# Create the plot
plt.figure(figsize=(8, 5))  # Set figure size
plt.plot(x, y, label='y = x²', color='b', linewidth=2)

# Customize the plot
plt.title("Line Plot of y = x²", fontsize=14)  # Add title
plt.xlabel("X Values", fontsize=12)  # Label for x-axis
plt.ylabel("Y Values (x²)", fontsize=12)  # Label for y-axis
plt.grid(True, linestyle='--', alpha=0.6)  # Add grid with dashed lines
plt.legend()  # Show legend

# Show the plot
plt.show()
```
