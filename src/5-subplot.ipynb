### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Data for plots
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
bar_colors = ['red', 'blue', 'green', 'purple']

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
pie_colors = ['blue', 'red', 'green', 'orange']
explode = (0.1, 0, 0, 0)

# 1st subplot: Line chart
axes[0, 0].plot(x, y, marker='o', linestyle='-', color='blue')
axes[0, 0].set_title("Line Chart")
axes[0, 0].set_xlabel("X-axis")
axes[0, 0].set_ylabel("Y-axis")
axes[0, 0].grid(True)

# 2nd subplot: Bar chart
axes[0, 1].bar(categories, values, color=bar_colors)
axes[0, 1].set_title("Bar Chart")
axes[0, 1].set_xlabel("Categories")
axes[0, 1].set_ylabel("Values")
axes[0, 1].grid(axis='y', linestyle='--', alpha=0.6)

# 3rd subplot: Scatter plot
axes[1, 0].scatter(x, y, color='blue', marker='o', edgecolors='black')
axes[1, 0].set_title("Scatter Plot")
axes[1, 0].set_xlabel("X Values")
axes[1, 0].set_ylabel("Y Values")
axes[1, 0].grid(True, linestyle='--', alpha=0.6)

# 4th subplot: Pie chart
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=pie_colors, 
               startangle=140, explode=explode, shadow=True)
axes[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()
plt.show()
```
